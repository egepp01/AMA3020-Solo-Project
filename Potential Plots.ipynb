```python
import numpy as np
import matplotlib.pyplot as plt

# Potential Plots
v = 2
R = 1
mu = [0, 0.5, 1, 2, 3, 5]
r = np.arange(0, 4, 0.01)

for mu in mu:
    U = (-2*v*(R**2)) / ((r**2) * (((r/R)**mu) + ((R/r)**mu))**2)
    plt.plot(r, U, label=f"$\mu={mu}$")
    
plt.legend(fontsize=16)
plt.ylim(-5,1)
plt.xlabel("r", fontsize=16)
plt.ylabel("$U(r)$", fontsize=16)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.title("U(r) for Different $\mu$", fontsize=16)
plt.savefig('potential plot')
plt.show()
```
